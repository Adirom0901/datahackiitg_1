```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/training_set_features.csv')
label = pd.read_csv('/content/training_set_labels.csv')
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe(include=['float64'])
```

### data cleaning


```python
df.xyz_concern =df.xyz_concern.fillna(2.0)
```

```python
df.xyz_knowledge = df.xyz_knowledge.fillna(1.0)
```

```python
# filling with most common class
df_clean = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
df_clean.head()
```

```python
df_clean.age_group.replace({'18 - 34 Years':0,'35 - 44 Years':1,'45 - 54 Years':2,'55 - 64 Years':3,'65+ Years':4},inplace=True)
```

```python
df_clean.education.replace({'< 12 Years':0,'12 Years':1,'College Graduate':3, 'Some College':2},inplace=True)
```

```python
df_clean.race.replace({'White':3,'Black':2,'Hispanic':1,'Other or Multiple':0},inplace=True)
```

```python
df_clean.income_poverty.replace({'Below Poverty':0,'<= $75,000, Above Poverty':1,'> $75,000':2},inplace=True)
```

```python
df_clean.marital_status.replace({'Married':1,'Not Married':0},inplace=True)
```

```python
df_clean.rent_or_own.replace({'Rent':0,'Own':1},inplace=True)
```

```python
df_clean.employment_status.replace({'Employed':2,'Unemployed':0,'Not in Labor Force':1},inplace=True)
```

```python
df_clean.census_msa.replace({'Non-MSA':0,'MSA, Not Principle  City':1,'MSA, Principle City':2},inplace=True)
```

```python
one_hot_encoded_data = pd.get_dummies(df_clean, columns =['sex'])
df_clean  = one_hot_encoded_data
```

```python
df_clean = pd.DataFrame(df_clean)
```

```python
df_clean = df_clean.drop(columns='employment_industry')
```

```python
df_clean = df_clean.drop(columns='employment_occupation')
```

```python
df_clean = df_clean.drop(columns='hhs_geo_region')
```

```python
df_clean = df_clean.astype(float)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_clean = scaler.fit_transform(df_clean)
```

```python
df_clean = pd.DataFrame(df_clean)
```

### splitting

```python
X_train = df_clean.iloc[:21000,:]
X_dev = df_clean.iloc[21000:, :]
y_train = label.iloc[:21000,:]
y_dev = label.iloc[21000:, :]
```

```python
y_dev = y_dev.drop(columns='respondent_id')
```

```python
y_train = y_train.drop(columns='respondent_id')
```

```python
y_train_1 = y_train.iloc[:,0]
y_train_2 = y_train.iloc[:,1]

y_dev_1 = y_dev.iloc[:,0]
y_dev_2 = y_dev.iloc[:,1]
```

```python
y_dev_1.shape
```

### model fit and evaluation

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

```python
xgb = GradientBoostingClassifier()
xgb.fit(X_train,y_train_1)
y_pred_xgb = xgb.predict_proba(X_dev)

rfr = RandomForestClassifier()
rfr.fit(X_train,y_train_1)
y_pred_rfr = rfr.predict_proba(X_dev)


lgr = LogisticRegression()
lgr.fit(X_train,y_train_1)
y_pred_lgr = lgr.predict_proba(X_dev)


nb = GaussianNB()
nb.fit(X_train,y_train_1)
y_pred_nb = nb.predict_proba(X_dev)
```

```python
y_pred_xgb_bin  = []
for  i in range(5707):
  y_pred_xgb_bin.append(np.argmax(y_pred_xgb[i,:]))
```

```python
y_pred_rfr_bin  = []
for  i in range(5707):
  y_pred_rfr_bin.append(np.argmax(y_pred_rfr[i,:]))
```

```python
y_pred_lgr_bin  = []
for  i in range(5707):
  y_pred_lgr_bin.append(np.argmax(y_pred_lgr[i,:]))
```

```python
y_pred_nb_bin  = []
for  i in range(5707):
  y_pred_nb_bin.append(np.argmax(y_pred_nb[i,:]))
```

```python
print('xgb',roc_auc_score(y_dev_1,y_pred_xgb_bin))

print('rfr',roc_auc_score(y_dev_1,y_pred_rfr_bin))


print('lgr',roc_auc_score(y_dev_1,y_pred_lgr_bin))


print('nb',roc_auc_score(y_dev_1,y_pred_nb_bin))
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(xgb, df_clean, label.iloc[:,1], cv=5))
```

```python
# Import necessary libraries
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for GridSearchCV
param_grid = {
	'n_estimators': [50, 100, 200],
	'learning_rate': [0.01, 0.1, 0.2],
	'max_depth': [3, 5, 7],
}

# Initialize the Gradient Boosting model
gb_model = GradientBoostingClassifier()

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=gb_model, param_grid=param_grid, cv=2, scoring='accuracy', n_jobs=-1)

# Fit the model to the training data using GridSearchCV
grid_search.fit(X_train, y_train_1)

# Get the best parameters and best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Make predictions on the test set using the best model
y_pred_best = best_model.predict(X_dev)

# Evaluate the best model
accuracy_best = accuracy_score(y_dev_1, y_pred_best)

# Print the results
print("Best Parameters:", best_params)
print(f"Best Model Accuracy: {accuracy_best}")
```
